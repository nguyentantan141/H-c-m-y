# Bài tập lý thuyết

**Bài 1: Tạo ngẫu nhiên $x,y \in R^{10}$**

```python
x=np.random.randint(0,9,size=10).reshape(1,10)
y=np.random.randint(0,9,size=10).reshape(1,10)
print('Vector x: ',x)
print('Vector y: ',y)
```

```python
# Cộng 2 vector
print('Cộng hai vector x,y: ',x+y)
# Trừ 2 vector
print('Trừ hai vector x,y: ', x-y)
# Nhân 2 vector
print('Nhân hai vector x,y: ',x*y)
# Chuyển vị ma trận x
print('Chuyển vị vector x:\n', x.T)
```

```python
# Độ dài vector
print("Độ dài vector x: ",np.linalg.norm(x))
print("Độ dài vector y: ",np.linalg.norm(y))
print("Độ dài vector (x+y): ",np.linalg.norm(x+y))
print("Độ dài vector (x-y): ",np.linalg.norm(x-y))
```

**Bài 2: Tạo ngẫu nhiên $X\in R^{5.7}$**

```python
# Tạo ma trận X:
matrix= np.random.randint(9,size=35).reshape(5,7)
print('Ma trận 5x7:\n',matrix)
print('Ma trận chuyển vị:\n',matrix.T)
```

```python
# Rank của ma trận
print('Rank của ma trận: ', np.linalg.matrix_rank(matrix))
```

**Hiển thị ảnh theo từng kênh màu**

```python
def display_im(i,j,image):  #i,j: kênh màu
  imcol=image.copy()
  imcol[:,:,i]=0
  imcol[:,:,j]=0
  cv2_imshow(imcol)
  print("nTensor của bức ảnh",imcol)
```

```python
display_im(0,1,im)
```

```python
display_im(1,2,im)
```

```python
display_im(2,0,im)
```

**Chuyển thư mục ảnh thành tensor 2 chiều**

```python
# Chuyển ảnh thành màu xám
def convert_gray(pathdic):  #pathdic: đường dẫn đến thư mục ảnh
  path = glob.glob(pathdic)
  for img in path:
    imgray = cv2.imread(img,0)
    vec=imgray.flatten()
    tensor_image=[]
    tensor_image.append(vec)
    tensor_image=np.array(tensor_image)
  return tensor_image
```

```python
print('Thư mục ảnh chuyển thành tensor:\n')
convert_gray('/content/drive/MyDrive/image/dataimage/Downloadimage/*.jpg')
```
