# Xử lý một văn bản

## Đọc dữ liệu

```python
def readData(pathfolder,mode: str):  # path: đường dẫn đến dữ liệu , mode: chế độ mở tệp 
  path= glob.glob(pathfolder)
  list_data=[]
  list_dic_document=[]
  for _ in path:
    with open(_, mode ) as f:
      file=f.read()
      listtext=file.split()
      list_data.append(listtext)  # Danh sách chứa các phần tử là các list chữ của từng văn bản
      settext=set(listtext)
      list_dic_document.append(settext)   # Danh sách chứa các phần tử là các từ điển của từng văn bản
  return  list_data, list_dic_document
```

```python
# Tạo từ điển cho thư mục 
def create_dictionary_folder(dic_folder,dic_):  # dic_ : từ điển từ của từng văn bản
  for _ in dic_:
    if _ not in  dic_folder:
      dic_folder.append(_)
  return dic_folder
```

```python
list_data,list_dic_document=readData('/content/drive/MyDrive/Data/tieng Anh/*.txt','rb')
dic_folder=[]
for _ in list_dic_document:
  dic_folder=create_dictionary_folder(dic_folder,_)

vector_freq=[]
freq = [(item, list_data[1].count(item)) for item in dic_folder]
for i in freq:
  print(i[1])
  vector_freq.append(i[1])
```

```python
def count_frequency_word(data,dic_folder):
  vector_freq=[]
  freq = [(item, data.count(item)) for item in dic_folder]
  for i in freq:
    vector_freq.append(i[1])
  return vector_freq
```

```python
def calculate_tf(data,dic_folder):
  vector_tf=[]
  vector_frequency=count_frequency_word(data,dic_folder)
  for j in vector_frequency:
    x=j/np.sum(vector_frequency)
    vector_tf.append(x)
  vector_tf=np.array(vector_tf)
  return vector_tf
```

```python
def calculate_idf(list_dic_document,dic_folder):
  n=len(list_dic_document)
  vector_idf=[]
  for _ in dic_folder:
    x=0
    for j in list_dic_document:
      if _ in j:
        x+=1
    idf=np.log((n/(1+x)))
    vector_idf.append(idf)
  vector_idf=np.array(vector_idf)
  return vector_idf  
```

```python
def calculate_tf_idf(vector_tf,vector_idf):
  vector_tf_idf=np.array(vector_tf)*np.array(vector_idf)
  return vector_tf_idf

```

```python
def main():
  list_data,list_dic_document=readData('/content/drive/MyDrive/Data/tieng Anh/*.txt','rb')
  dic_folder=[]
  list_freq=[]
  list_vector_tf=[]
  list_vector_tf_idf=[]
  for _ in list_dic_document:
    dic_folder=create_dictionary_folder(dic_folder,_)
  for _ in list_data:
    fre=count_frequency_word(_,dic_folder)
    list_freq.append(fre)
  for _ in list_freq:
    vector_tf=calculate_tf(_,dic_folder)
    list_vector_tf.append(vector_tf)
  vector_idf=calculate_idf(list_dic_document,dic_folder)
  for _ in list_vector_tf:
    vector_tf_idf=calculate_tf_idf(_,vector_idf)
    list_vector_tf_idf.append(vector_tf_idf)
  matrix_tf_idf=np.array(list_vector_tf_idf)
  vector_data=matrix_tf_idf.flatten()
  vector_data=vector_data.reshape((len(vector_data),1))
  print(vector_data)
  return vector_data




```

```python
if __name__ == "__main__":
    main()
```
